```python
import requests
from bs4 import BeautifulSoup as soup
```

```python
BASE_URL = "http://www.moh.gov.my/english.php/edirectory/member_list/12"
headers = {'user-agent':"Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/67.0.3396.87 Safari/537.36"}
req = requests.get(BASE_URL, headers=headers)
```

```python
if int(req.status_code)==200:
    html_soup = soup(req.content, "lxml")
    try:
        container = html_soup.find('div', {'id':'paginationdiv'})
    except:
        pass
    
            
    
```

```python
import csv
# file name 
fieldnames = ['Name ', 'Phone no. ','Email ','Division ', 'Section ']
csv_file_name = "Ministry Of Health Malayasia.csv"
def csv_file():
    with open(csv_file_name, 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()

# CSV file write row
def save_to_csv(data):
    with open(csv_file_name, 'a') as csvfile:
        writer = csv.writer(csvfile)
        print (data)
        writer.writerow(data.values())

csv_file()
page = container.findChild('div')
while page:
    list_items = page.findAll('div',{'class':'list-item'})
    for item in list_items:
        profile = item.find('div',{'class':'profile-detail'})
        name = profile.h1.text
        contents = profile.findAll('div', {'class':'profile-content'})
        data = {}
        data['Name '] = name
        for content in contents:
            content = content.text.split(':')
            data[content[0].replace('\n','')] = content[1].replace('\n','')
        save_to_csv(data)
    page = page.findNextSibling()
        
```
